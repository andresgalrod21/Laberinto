```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# ====================== SEGMENTO 1: Generación del laberinto ======================

def generar_laberinto(tamaño=100, densidad_obstaculos=0.3):
    # Crear un laberinto como una matriz de ceros
    laberinto = np.zeros((tamaño, tamaño), dtype=int)
    # Generar obstáculos aleatorios basados en la densidad especificada
    laberinto[np.random.rand(*laberinto.shape) < densidad_obstaculos] = 1
    # Asegurar que el punto inicial (0,0) esté libre
    laberinto[0, 0] = 0
    # Colocar la meta en una posición específica (5, 5)
    laberinto[99, 99] = 2
    return laberinto

# Generar el laberinto
laberinto = generar_laberinto(100, 0.3)

# Mostrar el laberinto generado gráficamente
plt.imshow(laberinto, cmap='viridis')  # Representar el laberinto con colores
plt.title("Laberinto Generado")
plt.show()

# ====================== SEGMENTO 2: Entrenamiento con Q-Learning ======================

# Definir las acciones posibles (arriba, abajo, izquierda, derecha) con sus desplazamientos
acciones = {0: (-1, 0), 1: (1, 0), 2: (0, -1), 3: (0, 1)}

def es_valido(laberinto, estado, accion):
    # Verificar si un movimiento es válido (dentro de límites y no hacia un obstáculo)
    filas, columnas = laberinto.shape
    movimiento = acciones[accion]
    nuevo_estado = (estado[0] + movimiento[0], estado[1] + movimiento[1])
    return (
        0 <= nuevo_estado[0] < filas and  # Dentro de las filas
        0 <= nuevo_estado[1] < columnas and  # Dentro de las columnas
        laberinto[nuevo_estado] != 1  # No es un obstáculo
    )

def q_learning(laberinto, episodios=1000, alpha=0.1, gamma=0.9, epsilon=0.1, max_movimientos=1000):
    # Crear una tabla Q inicializada en ceros
    q_table = np.zeros((laberinto.shape[0], laberinto.shape[1], len(acciones)))
    # Localizar las coordenadas de la meta
    meta = tuple(np.argwhere(laberinto == 2)[0])

    for episodio in range(episodios):  # Entrenar por el número de episodios especificados
        estado = (0, 0)  # Punto inicial
        movimientos = 0  # Contador de movimientos

        while estado != meta and movimientos < max_movimientos:
            fila, columna = estado
            # Elegir acción: exploración (aleatoria) o explotación (mejor valor Q)
            if np.random.rand() < epsilon:
                accion = np.random.choice(list(acciones.keys()))  # Exploración
            else:
                accion = np.argmax(q_table[fila, columna])  # Explotación

            # Validar si el movimiento es válido
            if es_valido(laberinto, estado, accion):
                movimiento = acciones[accion]
                nuevo_estado = (fila + movimiento[0], columna + movimiento[1])
                recompensa = 1 if nuevo_estado == meta else -0.1  # Recompensa
                # Actualizar la tabla Q con la fórmula del Q-Learning
                q_table[fila, columna, accion] += alpha * (
                    recompensa + gamma * np.max(q_table[nuevo_estado]) - q_table[fila, columna, accion]
                )
                estado = nuevo_estado  # Actualizar al nuevo estado

            movimientos += 1  # Incrementar el número de movimientos

        # Mostrar progreso cada 100 episodios
        if episodio % 100 == 0:
            print(f"Episodio {episodio}/{episodios} completado.")

    return q_table

# Entrenar el agente y generar la tabla Q
q_table = q_learning(laberinto, episodios=1000, alpha=0.1, gamma=0.9, epsilon=0.1)

# ====================== SEGMENTO 3: Evaluación y visualización ======================

def probar_agente(laberinto, q_table):
    estado = (0, 0)  # Punto inicial
    meta = tuple(np.argwhere(laberinto == 2)[0])  # Coordenadas de la meta
    camino = [estado]  # Lista para almacenar el camino recorrido

    while estado != meta:  # Mientras no se alcance la meta
        fila, columna = estado
        accion = None

        # Elegir la mejor acción válida basada en la tabla Q
        for accion_candidata in np.argsort(-q_table[fila, columna]):  # Ordenar acciones por valor Q
            if es_valido(laberinto, estado, accion_candidata):  # Verificar si la acción es válida
                accion = accion_candidata
                break

        # Si no hay acciones válidas, finalizar
        if accion is None:
            print("No hay acciones válidas disponibles, terminando evaluación.")
            break

        # Realizar el movimiento y actualizar el estado
        movimiento = acciones[accion]
        nuevo_estado = (fila + movimiento[0], columna + movimiento[1])
        estado = nuevo_estado
        camino.append(estado)  # Registrar el nuevo estado en el camino

    return camino

# Evaluar el agente y obtener el camino aprendido
camino = probar_agente(laberinto, q_table)

def mostrar_camino(laberinto, camino):
    # Copiar el laberinto para marcar el camino
    laberinto_copia = laberinto.copy()
    for (fila, columna) in camino:
        if laberinto_copia[fila, columna] == 0:  # Solo marcar celdas transitables
            laberinto_copia[fila, columna] = 3

    # Crear un mapa de colores personalizado para mostrar el camino
    cmap = ListedColormap(["black", "gray", "red", "white"])  # Fondo, obstáculos, meta, camino
    plt.imshow(laberinto_copia, cmap=cmap)  # Mostrar el laberinto con el camino aprendido
    plt.title("Camino Aprendido")
    plt.show()

# Mostrar gráficamente el camino aprendido
mostrar_camino(laberinto, camino)
```
